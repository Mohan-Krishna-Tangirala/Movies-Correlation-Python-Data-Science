```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Movies.csv')
df.head()
```

### Data Cleaning

```python
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['budget'], inplace = True)
df.dropna(subset=['gross'], inplace = True)
```

```python
df.isnull().sum()
```

```python
# changing the data type of columns budget and gross from float to int

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
df.duplicated().sum()
```

### Correlation

```python
# Scatter plot / budget vs gross

sns.scatterplot(data = df, x = 'budget', y = 'gross')
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')

plt.show()
```

```python
sns.regplot(data = df, x = 'budget', y = 'gross')
plt.show()
```

```python
df.corr()
# high correaltion between budget and gross
```

```python
sns.heatmap(df.corr(), annot = True)
plt.title('Correlation matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# changing all the columns to numericals
```

```python
df1 = df.copy()

for col_name in df1.columns:
    if(df1[col_name].dtype == 'object'):
        df1[col_name] = df1[col_name].astype('category')
        df1[col_name] = df1[col_name].cat.codes
        
df1
        
```

```python
df
```

```python
sns.heatmap(df1.corr(), annot = True)
plt.title('Correlation matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df1.corr()
```

```python
# let's see highest correlation
```

```python
corr_matrix = df1.corr()

corr_pairs =  corr_matrix.unstack()
corr_pairs
```
